```python

def director(obj):
    l = []
    if isinstance(obj, str):
        obj = ast.literal_eval(obj)
    for i in obj:
        if i['job'] == 'Director':
            l.append(i['name'])
    return l

```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
    
```
